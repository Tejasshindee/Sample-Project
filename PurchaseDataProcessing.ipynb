# Remove numeric values from category


```python


# Convert non-numeric values in 'category' column to string
dfcf['category_code'] = dfcf['category_code'].astype(str)

# # Remove rows where 'category' column contains numeric values
non_numeric_df = dfcf[~dfcf['category_code'].str.isnumeric()]

# print("DataFrame after removing rows with numeric values in 'category' column:")
print(non_numeric_df)
```

# Removing numertic columns in category_code column to excel

```python
non_numeric_df.to_excel('cleaned_file_2.xlsx', index=False)
```

# Check 'nan' values in category_code column

```python
num_other_occurrences = (non_numeric_df['category_code'] == 'nan').sum()

print("Number of occurrences of 'nan' in the 'category' column:", num_other_occurrences)
```

# Assign 'other' value to 'nan' value in category_code

```python
non_numeric_df.loc[non_numeric_df['category_code'] == 'nan', 'category_code'] = 'other'

print("DataFrame after assigning 'other' to all occurrences of 'nan' in the 'category_code' column:")
print(non_numeric_df)
```

# Chech NaN values in price column

```python
num_nan_values = non_numeric_df['price'].isna().sum()

print("Number of NaN values in the 'price' column:", num_nan_values)
```

# Remove NaN values from price column

```python
rem_nul_price = non_numeric_df[~non_numeric_df['price'].isnull()]
rem_nul_price
```

# Removing records that doesn't contain price


```python
rem_nul_price.to_excel('cleaned_file_3.xlsx', index=False)
```

```python
num_nan_values = rem_nul_price['price'].isna().sum()

print("Number of NaN values in the 'price' column:", num_nan_values)
```

# Assign 'other' to 'nan' value in category_code (after removing null price)
