# Remove numeric values from category


```python


# Convert non-numeric values in 'category' column to string
dfcf['category_code'] = dfcf['category_code'].astype(str)

# # Remove rows where 'category' column contains numeric values
non_numeric_df = dfcf[~dfcf['category_code'].str.isnumeric()]

# print("DataFrame after removing rows with numeric values in 'category' column:")
print(non_numeric_df)
```

# Removing numertic columns in category_code column to excel

```python
non_numeric_df.to_excel('cleaned_file_2.xlsx', index=False)
```

# Check 'nan' values in category_code column

```python
num_other_occurrences = (non_numeric_df['category_code'] == 'nan').sum()

print("Number of occurrences of 'nan' in the 'category' column:", num_other_occurrences)
```

# Assign 'other' value to 'nan' value in category_code

```python
non_numeric_df.loc[non_numeric_df['category_code'] == 'nan', 'category_code'] = 'other'

print("DataFrame after assigning 'other' to all occurrences of 'nan' in the 'category_code' column:")
print(non_numeric_df)
```

# Count Number of INCONSISTENT records of order_id and event_time

```python
# Group by order_id and check for inconsistencies
inconsistent_order_ids = []
for order_id, group in rem_nul_brand.groupby('order_id'):
    if group['event_time'].nunique() > 1:
        inconsistent_order_ids.append(order_id)

# Count the number of records with inconsistent event_time
num_inconsistent_records = rem_nul_brand[rem_nul_brand['order_id'].isin(inconsistent_order_ids)].shape[0]

print("Number of records with inconsistent event_time:", num_inconsistent_records)
```

# Display inconsistent records

```python
inconsistent_order_ids = []
for order_id, group in rem_nul_brand.groupby('order_id'):
    if group['event_time'].nunique() > 1:
        inconsistent_order_ids.append(order_id)

# Filter DataFrame to include only rows with inconsistent event_time
inconsistent_rows = rem_nul_brand[rem_nul_brand['order_id'].isin(inconsistent_order_ids)]

print("Rows with inconsistent event_time:")
print(inconsistent_rows)
```

# Removes inconsistent event_time records

```python
# # Remove records with inconsistent event_time
df_cleaned = rem_nul_brand[~rem_nul_brand['order_id'].isin(inconsistent_order_ids)]
print("DataFrame after removing records with inconsistent event_time:")
print(df_cleaned)
```
